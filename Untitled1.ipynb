```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
```

```python

# Step 1: Load the datasets
train_main = pd.read_csv('Train-1542865627584.csv')
train_beneficiary = pd.read_csv('Train_Beneficiarydata-1542865627584.csv')
```

```python
# Step 2: Concatenate the dataframes side-by-side
kf= pd.concat([train_beneficiary, train_main], axis=1)
```

```python
# Concatenate the DataFrame with itself to double the rows
df = pd.concat([kf, kf], ignore_index=True)
```

```python
df
```

```python
df.isnull()
```

```python
# Step 1: Replace 'Unknown' with random 1 or 0
df['PotentialFraud'] = df['PotentialFraud'].apply(lambda x: np.random.choice([0, 1]) if x == 'Unknown' else x)
```

```python
# Step 2: If 'PotentialFraud' column contains non-numeric values like 'No' or 'Yes', convert them to integers
df['PotentialFraud'] = df['PotentialFraud'].apply(lambda x: 1 if x == 'Yes' else (0 if x == 'No' else x))
```

```python
# Step 3: Convert the column to numeric, forcing errors to NaN (if there are any left)
df['PotentialFraud'] = pd.to_numeric(df['PotentialFraud'], errors='coerce')
```

```python
# Optional: Fill any remaining NaN values with a default value (e.g., 0)
df['PotentialFraud'] = df['PotentialFraud'].fillna(0)
```

```python
from sklearn.utils import resample
```

```python
df
```

```python
# Step 4: Handle 'Provider' column
df['Provider'] = df['Provider'].fillna('Unknown')
df_filtered = df[df['Provider'] != 'Unknown']['Provider'].astype(str)
```

```python
# Generate sequential provider codes for 'Unknown' entries
if len(df_filtered) > 0:
    last_provider = df_filtered.max()  
    last_number = int(last_provider.replace('PRV', '')) if last_provider.startswith('PRV') else 0
    unknown_count = df['Provider'].eq('Unknown').sum()
    new_provider_codes = [f'PRV{i}' for i in range(last_number + 1, last_number + 1 + unknown_count)]
    df.loc[df['Provider'] == 'Unknown', 'Provider'] = new_provider_codes
```

```python
df['Provider']
```

```python
# Drop the  column
df = df.drop('Unnamed: 25', axis=1)
```

```python
# Step 5: Encode categorical features
label_encoder = LabelEncoder()
df['BeneID'] = label_encoder.fit_transform(df['BeneID'])
df['DOB'] = label_encoder.fit_transform(df['DOB'])
df['DOD'] = label_encoder.fit_transform(df['DOD'])
df['Provider'] = label_encoder.fit_transform(df['Provider'])
```

```python

# Step 6: Split data into features (X) and target (y)
X = df.drop('PotentialFraud', axis=1)
y = df['PotentialFraud']
```

```python
y
```

```python
X
```

```python

# Step 7: Convert all non-numeric values to NaN, then fill NaNs with median
X = X.apply(pd.to_numeric, errors='coerce')
X = X.fillna(X.median())
```

```python

# Step 8: Handle class imbalance with SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Step 9: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)
```

```python

# Step 10: Train an advanced model (XGBoost)
model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)
```

```python
model.fit(X_train, y_train)
```

```python
# Step 11: Evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Model accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import f1_score, confusion_matrix
```

```python
# Calculate the F1 Score
f1 = f1_score(y_test, y_pred, average='binary')
print("F1 Score:", f1)
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Step 7: Plot confusion matrix using Seaborn heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['No Fraud', 'Fraud'], yticklabels=['No Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Optional: Plot a bar chart for F1 score comparison (e.g., between models)
plt.bar(['F1 Score'], [f1], color='blue')
plt.ylabel('Score')
plt.title('F1 Score for the Model')
plt.show()
```

```python
import pandas as pd
import pickle
from sklearn.impute import SimpleImputer

# Load the trained model (replace with the actual model file path)
with open('xgboost_model.pkl', 'rb') as file:
    model = pickle.load(file)

# Initialize the SimpleImputer to handle missing values (using mean strategy)
imputer = SimpleImputer(strategy='mean')

# Function to get numeric user input with examples
def get_user_input():
    print("Please enter the following details:")
    user_input = {}

    # Feature names and their expected types
    feature_types = {
        'BeneID': 'int', 'DOB': 'int', 'DOD': 'int', 'Gender': 'int', 'Race': 'int', 
        'RenalDiseaseIndicator': 'float', 'State': 'int', 'County': 'int', 
        'NoOfMonths_PartACov': 'int', 'NoOfMonths_PartBCov': 'int', 
        'ChronicCond_Alzheimer': 'int', 'ChronicCond_Heartfailure': 'int',
        'ChronicCond_KidneyDisease': 'int', 'ChronicCond_Cancer': 'int',
        'ChronicCond_ObstrPulmonary': 'int', 'ChronicCond_Depression': 'int',
        'ChronicCond_Diabetes': 'int', 'ChronicCond_IschemicHeart': 'int',
        'ChronicCond_Osteoporasis': 'int', 'ChronicCond_rheumatoidarthritis': 'int',
        'ChronicCond_stroke': 'int', 'IPAnnualReimbursementAmt': 'int', 
        'IPAnnualDeductibleAmt': 'int', 'OPAnnualReimbursementAmt': 'int',
        'OPAnnualDeductibleAmt': 'int', 'Provider': 'int'
    }

    # Loop to get inputs from the user
    for feature, feature_type in feature_types.items():
        while True:
            try:
                if feature_type == 'int':
                    # Integer input for features, including binary features (1 for No, 2 for Yes)
                    value = input(f"Enter {feature} (integer, e.g., 12345; use 1 for No and 2 for Yes for binary fields): ")
                    user_input[feature] = int(value)
                elif feature_type == 'float':
                    # Float input for specific features
                    value = input(f"Enter {feature} (float value, e.g., 3.0): ")
                    user_input[feature] = float(value)

                break  # Exit the loop if valid input is entered
            except ValueError:
                print(f"Invalid input for {feature}. Please enter a valid {feature_type} value.")

    return user_input

# Function to process the user input (convert to DataFrame and handle missing values)
def process_input(user_input):
    # Convert the user input dictionary to a DataFrame
    user_input_df = pd.DataFrame([user_input])

    # Apply imputation to handle missing values (filling with mean for each column)
    user_input_df = pd.DataFrame(imputer.fit_transform(user_input_df), columns=user_input_df.columns)
    
    return user_input_df

# Function to make a prediction based on user input
def predict_fraud(user_input_df):
    prediction = model.predict(user_input_df)

    # Return the prediction as either 'Fraud' or 'Not Fraud'
    if prediction[0] == 1:
        return "The input data is predicted as 'Fraud'."
    else:
        return "The input data is predicted as 'Not Fraud'."

# Main function to run the user input and prediction process
def main():
    print("Welcome! Let's collect some information and predict if it's fraudulent or not.")
    
    # Step 1: Get user input
    user_input = get_user_input()
    
    # Step 2: Process the user input
    user_input_df = process_input(user_input)
    
    # Step 3: Make the prediction
    result = predict_fraud(user_input_df)
    
    # Step 4: Print the result
    print(result)

# Run the program
if __name__ == '__main__':
    main()
```
